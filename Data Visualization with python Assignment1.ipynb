```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = [12,9]
```

Read the companies file into Pandas data frame and print a list of companies based in the USA

```python
comp = pd.read_csv('companies.csv')
comp.head()
```

```python
comp_usa = comp[comp['country']=='USA']
comp_usa
```

Sort the companies in descending order of revenue and generate a bar chart

```python
plt.title('Companies in descending order of revenue')
sns.barplot(x= 'annualRevenue', 
            y = 'companyName', 
            data = comp, 
            order = comp.sort_values('annualRevenue', ascending = False).companyName,
            orient = 'h')
plt.show()
```

Generate a bubble chart showing companies by the number of employees

```python
plt.title('Companies by the number of employees')
sns.scatterplot(data = comp,
                x = 'founded', 
                y = 'companyName',
                size = 'employees', 
                hue = 'companyName',
                legend = False, 
                sizes = (3000,20000))
plt.show()
```

Generate two pie charts of companies by revenue and by the number of employees

```python
plt.pie(x = 'annualRevenue',
        data = comp, 
        labels = 'companyName', 
        autopct = '%.1f%%', 
        explode=[0.05]*len(comp['companyName']))
plt.title('Companies by revenue')
plt.show()
```

```python
plt.pie(x = 'employees', 
        data = comp, 
        labels = 'companyName', 
        autopct = '%.1f%%', 
        explode=[0.05]*len(comp['companyName']))
plt.title('Companies by the number of employees')
plt.show()
```
